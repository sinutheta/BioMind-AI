```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
X_diab = sugar_df.drop("Outcome", axis=1)
y_diab = sugar_df["Outcome"]

X_d_train, X_d_test, y_d_train, y_d_test = train_test_split( X_diab, y_diab, test_size=0.2, random_state=42, stratify=y_diab)

print("Diabetes train:", X_d_train.shape)
print("Diabetes test:", X_d_test.shape)
```

```python
zero_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
X_d_train[zero_cols] = X_d_train[zero_cols].replace(0, np.nan)
X_d_test[zero_cols]  = X_d_test[zero_cols].replace(0, np.nan)
train_medians = X_d_train.median()
X_d_train.fillna(train_medians, inplace=True)
X_d_test.fillna(train_medians, inplace=True)
print("Nulls remaining:", X_d_train.isnull().sum().sum())  
```

```python

Q1 = X_d_train["Insulin"].quantile(0.25)
Q3 = X_d_train["Insulin"].quantile(0.75)
upper = Q3 + 1.5 * (Q3 - Q1)
X_d_train["Insulin"] = X_d_train["Insulin"].clip(upper=upper)
X_d_test["Insulin"]  = X_d_test["Insulin"].clip(upper=upper)


scaler_diab = StandardScaler()
X_d_train = scaler_diab.fit_transform(X_d_train)
X_d_test  = scaler_diab.transform(X_d_test)

print("Diabetes scaled shape:", X_d_train.shape)
```

```python
sm = SMOTE(random_state=42)
X_d_train, y_d_train = sm.fit_resample(X_d_train, y_d_train)

print("After SMOTE:", pd.Series(y_d_train).value_counts())
```

```python
X_heart = heart_df.drop("target", axis=1)
y_heart = heart_df["target"]

X_h_train, X_h_test, y_h_train, y_h_test = train_test_split( X_heart, y_heart, test_size=0.2, random_state=42, stratify=y_heart)

cat_cols = ["cp", "restecg", "slope", "thal"]
X_h_train = pd.get_dummies(X_h_train, columns=cat_cols, drop_first=True)
X_h_test  = pd.get_dummies(X_h_test,  columns=cat_cols, drop_first=True)

X_h_test = X_h_test.reindex(columns=X_h_train.columns, fill_value=0)

Q3_chol    = X_h_train["chol"].quantile(0.75)
upper_chol = Q3_chol + 1.5 * (Q3_chol - X_h_train["chol"].quantile(0.25))
X_h_train["chol"] = X_h_train["chol"].clip(upper=upper_chol)
X_h_test["chol"]  = X_h_test["chol"].clip(upper=upper_chol)

scaler_heart = StandardScaler()
X_h_train = scaler_heart.fit_transform(X_h_train)
X_h_test  = scaler_heart.transform(X_h_test)

print("Heart train shape:", X_h_train.shape)
```

```python

# ── Load models ──────────────────────────────────────────────────
normal_model  = joblib.load("models/normal_model.pkl")
heart_model    = joblib.load("models/heart_model.pkl")
sugar_model = joblib.load("models/sugar_model.pkl")
scaler_heart   = joblib.load("models/scaler_heart.pkl")
scaler_diab    = joblib.load("models/scaler_diab.pkl")


# ── Helper: encode + scale heart input ──────────────────────────
def prepare_heart_input(raw_input: dict):
    # Step 1: Pull out categorical values
    cp       = raw_input["cp"]
    restecg  = raw_input["restecg"]
    slope    = raw_input["slope"]
    thal     = raw_input["thal"]

    # Step 2: Build base numeric row (non-categorical columns)
    row = {
        "age":      raw_input["age"],
        "sex":      raw_input["sex"],
        "trestbps": raw_input["trestbps"],
        "chol":     raw_input["chol"],
        "fbs":      raw_input["fbs"],
        "thalach":  raw_input["thalach"],
        "exang":    raw_input["exang"],
        "oldpeak":  raw_input["oldpeak"],
        "ca":       raw_input["ca"],
    }

    # Step 3: Manually create dummy columns (drop_first=True drops _0 for each)
    for val in [1, 2, 3]:
        row[f"cp_{val}"]      = 1 if cp      == val else 0
    for val in [1, 2]:
        row[f"restecg_{val}"] = 1 if restecg == val else 0
    for val in [1, 2]:
        row[f"slope_{val}"]   = 1 if slope   == val else 0
    for val in [1, 2, 3]:
        row[f"thal_{val}"]    = 1 if thal    == val else 0

    # Step 4: Build DataFrame and align to exact training columns
    df_input      = pd.DataFrame([row])
    train_columns = heart_model.get_booster().feature_names
    df_input      = df_input.reindex(columns=train_columns, fill_value=0)

    return scaler_heart.transform(df_input)


# ── Lifestyle multiplier ─────────────────────────────────────────
def compute_lifestyle_modifier(data: dict):
    m = 1.0
    breakdown = {}

    sleep = data.get("sleep_hours", 7)
    if sleep < 4:       d = +0.50
    elif sleep < 6:     d = +0.25
    elif sleep >= 8:    d = -0.10
    else:               d = 0.0
    m += d; breakdown["sleep"] = d

    stress = data.get("stress_level", 3)
    if stress >= 8:     d = +0.30
    elif stress >= 5:   d = +0.10
    else:               d = 0.0
    m += d; breakdown["stress"] = d

    ex = data.get("exercise_days", 3)
    if ex == 0:         d = +0.20
    elif ex >= 5:       d = -0.15
    else:               d = 0.0
    m += d; breakdown["exercise"] = d

    d = +0.40 if data.get("smoker") else 0.0
    m += d; breakdown["smoker"] = d

    d = +0.20 if data.get("alcohol_units_week", 0) > 14 else 0.0
    m += d; breakdown["alcohol"] = d

    return round(max(0.5, min(2.5, m)), 2), breakdown


# ── Main prediction function ─────────────────────────────────────
def predict_all(symptom_vector, heart_features: dict, diabetes_features, lifestyle_data: dict):

    # --- Model 1: Symptom ---
    sym_proba_matrix = normal_model.predict_proba([symptom_vector])
    sym_probs        = sym_proba_matrix[0]
    sym_classes      = normal_model.classes_

    class_prob_pairs = [(str(sym_classes[i]), float(sym_probs[i])) for i in range(len(sym_classes))]
    top3_sym         = sorted(class_prob_pairs, key=lambda x: -x[1])[:3]
    sym_scores       = {disease: round(prob, 3) for disease, prob in top3_sym}

    # --- Model 2a: Heart ---
    heart_scaled = prepare_heart_input(heart_features)
    heart_proba  = heart_model.predict_proba(heart_scaled)
    heart_risk   = round(float(heart_proba[0][1]), 3)

    # --- Model 2b: Diabetes ---
    diab_scaled = scaler_diab.transform([diabetes_features])
    diab_proba  = sugar_model.predict_proba(diab_scaled)
    diab_risk   = round(float(diab_proba[0][1]), 3)

    # --- Model 3: Lifestyle multiplier ---
    modifier, breakdown = compute_lifestyle_modifier(lifestyle_data)

    # --- Combine all scores ---
    base_scores  = {**sym_scores, "heart_disease": heart_risk, "diabetes": diab_risk}
    final_scores = {
        k: round(min(1.0, v * modifier), 3)
        for k, v in base_scores.items()
    }

    return {
        "final_scores": final_scores,
        "lifestyle_modifier": modifier,
        "lifestyle_breakdown": breakdown
    }


# ── Test with a sample patient ───────────────────────────────────
sym_vector = [0] * len(normal_model.feature_names_in_)

heart_input = {
    "age": 52, "sex": 1, "trestbps": 140, "chol": 230,
    "fbs": 0, "thalach": 150, "exang": 1, "oldpeak": 1.5, "ca": 0,
    "cp": 2, "restecg": 1, "slope": 2, "thal": 2
}

diabetes_input = [3, 140, 72, 35, 150, 33.0, 0.627, 45]

lifestyle_input = {
    "sleep_hours": 5,
    "stress_level": 7,
    "exercise_days": 1,
    "smoker": False,
    "alcohol_units_week": 5
}

# ── Run and print ────────────────────────────────────────────────
result = predict_all(sym_vector, heart_input, diabetes_input, lifestyle_input)

print("=== Final Scores ===")
for disease, score in result["final_scores"].items():
    print(f"  {disease}: {round(score * 100, 1)}%")

print(f"\nLifestyle Modifier: {result['lifestyle_modifier']}x")
print("\nLifestyle Breakdown:")
for factor, delta in result["lifestyle_breakdown"].items():
    if delta != 0:
        sign = "+" if delta > 0 else ""
        print(f"  {factor}: {sign}{delta}")
```
